```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix
```

```python
df = pd.read_csv("C:/Users/TiaSP/Downloads/newChurn_data.csv", index_col="customerID")
df
```

```python
#Checking that our data is in the proper format with no missing values.
df.info()
```

```python
#Breaking up data into features and targets
features = df.drop('Churn', axis=1)
targets = df['Churn']
```

```python
#Splitting data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(features, targets, stratify=targets, random_state=42)
```

```python
#Fitting logistic regression model to training data set
lr_model = LogisticRegression(max_iter=1000)
lr_model.fit(x_train, y_train)
```

```python
#Evaluating our model based on "no information" rate: The "no information" rate is 73.5%
df['Churn'].value_counts(normalize=True)
```

```python
#Evaluating our model based on score method 
print(lr_model.score(x_train, y_train))
print(lr_model.score(x_test, y_test))

#Both sets are moderately accurate.
#The train set has 79.7% accuracy while the test set has 78.4% accuracy.
#The train set has a slightly higher accuracy accuracy than the test set.
#The accuracies of both the train and test sets are higher than the model's "no information" rate of 73.5%.
```

```python
#Using a confusion matrix to evaluate our model
plot_confusion_matrix(lr_model, x_test, y_test, cmap='Blues')
```

```python
#Observing the coefficients of the model
lr_model.coef_
```

```python
#Viewing the columns that correspond to each coefficient
features.columns
```

```python
#Setting dataframe for coefficients
coef_df = pd.DataFrame(data=lr_model.coef_, columns=features.columns)
```

```python
#Generating bar plot for coefficients and variables
coef_df.T.sort_values(by=0).plot.bar(legend=False)
```
