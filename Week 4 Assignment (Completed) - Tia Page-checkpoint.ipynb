```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
#Creating data frame using last week's Churn data set
df = pd.read_csv("C:/Users/TiaSP/Downloads/newChurn_data.csv", index_col = "customerID")
df
```

```python
#Breaking our data into features and targets

features = df.drop('Churn', axis=1)
targets = df['Churn']
x_train, x_test, y_train, y_test = train_test_split(features, targets, stratify=targets, random_state=42)

```

```python
#Fitting decision tree to the training data; adjusted max_depth to 6 to reduce overfitting.

dt = DecisionTreeClassifier(max_depth=6)
dt.fit(x_train, y_train)

print(dt.score(x_train, y_train))
print(dt.score(x_test, y_test))
```

```python
#Seeing how far the decision tree goes
dt.get_depth()
```

```python
#Plotting decision tree
f = plt.figure(figsize=(25, 25))
_ = plot_tree(dt, fontsize=8, feature_names=features.columns, filled=True)
```

```python
#Importing seaborn package
import seaborn as sns
```

```python
#Importing phi-k package
import phik

df.phik_matrix()
```

```python
#Visualizing correlations through a heatmap
sns.heatmap(df.phik_matrix())
```

```python
#Fitting random forest model to Churn data set

from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(max_depth=5, max_features=7, random_state=42)
rfc.fit(x_train, y_train)

print(rfc.score(x_train, y_train))
print(rfc.score(x_test, y_test))
```

```python
#Plotting feature importances

from scikitplot.estimators import plot_feature_importances

plot_feature_importances(rfc, feature_names=features.columns, x_tick_rotation=90)
```

```python
#Eliminating the least important features
new_features = features.drop(['PhoneService', 'PaymentMethod', 'TotalCharges'], axis=1)
x_train, x_test, y_train, y_test = train_test_split(new_features, targets, stratify=targets, random_state=42)
```

```python
#Fitting random forest model with our new data

rfc = RandomForestClassifier(max_depth=7, max_features=4, random_state=42)
rfc.fit(x_train, y_train)

print(rfc.score(x_train, y_train))
print(rfc.score(x_test, y_test))
```

```python
#Examining feature importances after removing less importances
plot_feature_importances(rfc, feature_names=new_features.columns, x_tick_rotation=90)
```
