```python
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/TiaSP/Downloads/churn_data.csv")
df
```

```python
#Checking for outliers
sns.boxplot(data=df, orient = "h" )

#There doesn't appear to be any outliers.
```

```python
#Check for missing values
df.isna().sum()

#There are 11 missing values for the TotalCharges column.
```

```python
#TotalCharges data is skewed to the right, so I will replace the missing values with the median.
df['TotalCharges'].fillna(df['TotalCharges'].median(), inplace=True)
```

```python
#Converting categorical columns into numeric values
df['customerID'] = df.index + 1
df['PhoneService'] = df['PhoneService'].replace({'Yes': 1, 'No': 0})
df['Contract'] = df['Contract'].replace({'Month-to-month': 0, 'One year': 1, 'Two year': 2})
df['PaymentMethod'] = df['PaymentMethod'].replace({'Electronic check': 1, 'Mailed check': 2, 'Bank transfer (automatic)': 3, 'Credit card (automatic)': 4})
df['Churn'] = df['Churn'].replace({'Yes': 1, 'No': 0})
```

```python
#Checking that all columns have numerical values
df.info()
```

```python
#Combining columns
df['Number of Payments Made'] = round(df['TotalCharges'] / df['MonthlyCharges'])
df['Number of Payments Made']
```

```python
#Plotting Total Charges vs. Number of Payments Made

sns.scatterplot(data=df, x='Number of Payments Made', y="TotalCharges", s=5)
```

## Analysis

Judging from the box plots, there were no outliers in the data. However, there were 11 missing rows for the TotalCharges column. Based on the data being skewed (to the right), I decided to replace the missing TotalCharges values with the median of the total charges. I created a new column titled 'Number of Payments Made' that divides the total amount of charges by the monthly charges in order to see how many payments a customer has made. In terms of the relationship between the number of payments and the total charges, I predicted there would be a positive linear relationship between the two variables. I tested to see if this prediction was true by visualizing the data on a scatterplot. The scatterplot shows that there is a positive linear relationship between number of payments made and total charges. I assumed this would be the case because it would not make sense for the total charges to decrease as a customer keeps making payments. I noticed in the plot that some of the customers who make a lot of payments did not have as much total charges as those who made less payments. This could possible be due to them paying less in services; customers who pay a higher rate each month will accumulate higher total charges through less payments than customers who pay less.
